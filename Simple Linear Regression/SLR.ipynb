### step-3 R&D VS Profit model

```python
import seaborn as sns
sns.scatterplot(data=df,x='RND',y='PROFIT')
```

#PROFIT~RND

```python
X=df[['RND']]
Y=df[['PROFIT']]
```

```python
X.head()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,Y)
```
